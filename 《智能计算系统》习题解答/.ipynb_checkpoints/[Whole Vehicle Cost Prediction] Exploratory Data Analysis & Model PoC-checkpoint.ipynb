column `PUR PRICE` only has NULL value, so drop it.

```python
cleaned_dat = raw_dat.drop(columns=['XITEM', 'DESB', 'LastPurchPurPrice', 'LastPurchPurDate', 'VendorID', 'Datekey', 'TruckType', 'PUR PRICE'])
```

```python
cleaned_dat.head()
```

```python
plt.figure(figsize = (32,18))
corr = cleaned_dat.corr()
sns.heatmap(corr.T, square=True, annot=True)
```

```python
drop_cols = ['Material', 'REQTE', 'SpecPurchType', 'DESA', 'PurchType', 
             'Material cost(Local)', 'Material cost(Import)', 'Material-OH-Fix(Local)', 'Material-OH-Fix(Import)', 'Material-OH-Var(Import)',
             'Fkvoll(Direct Labour)', 'Fkvoll(Production-OH-F)', 'Fkvoll(Production-OH-V)',
             'LastPurchCurrency', 'Last Purchase Price', 'Type', 'Total Cost']

corr = cleaned_dat.drop(columns=drop_cols).corr()
plt.figure(figsize = (21,12))
sns.heatmap(corr.T, square=True, annot=True)
```

```python
sns.pairplot(cleaned_dat.drop(columns=drop_cols))
```

```python
cleaned_dat.drop(columns=drop_cols).shape
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(16,9))
i, j = 0, 0
for col in cleaned_dat.drop(columns=drop_cols).columns:
    with warnings.catch_warnings():
        warnings.simplefilter('ignore')
        sns.distplot(cleaned_dat.drop(columns=drop_cols)[col], ax=axes[i][j])
    j += 1
    if j == 2: i, j = i+1, 0
plt.tight_layout()
```
